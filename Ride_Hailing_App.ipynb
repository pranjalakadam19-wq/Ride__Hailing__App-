### Ride-Hailing App (Uber/Ola style)

Design a **basic Ride-Hailing System** using OOP in Python.

##### Requirements:

1. **Classes**

   * `Driver` → attributes: `name`, `car_model`, `rating`
   * `Rider` → attributes: `name`, `location`, `destination`
   * `Ride` → attributes: `driver`, `rider`, `fare`

2. **Methods**

   * In `Ride`, create a method `calculate_fare(distance)` that charges ₹10 per km.
   * In `Ride`, create a method `ride_info()` that displays rider name, driver name, car model, and fare.

3. **Task**

   * Create one driver and one rider.
   * Start a ride for 15 km.
   * Print ride details using `ride_info()`.

---

###  Expected Output

```
Rider: Amit | Driver: Rajesh (Toyota Innova) | Distance: 15 km | Fare: ₹150
```


````python
class Driver:
  def __init__(self,name,car_model,rating):
    self.name = name
    self.car_model = car_model
    self.rating = rating

class Rider:
  def __init__(self,name,location,destination):
    self.name = name
    self.location = location
    self.destionation = destination

class Ride:
  def __init__(self,drive,rider):
    self.drive = drive
    self.rider = rider
    self.fare = 0
    self.distance = 0

  def calculate_fare(self,distance):
    self.distance = distance
    self.fare = distance * 10
    return f"The total fare for the {distance} km is {self.fare} rupees"

  def ride_info(self):
    return f"Rider: {self.rider.name} | Driver: {self.drive.name} {self.drive.car_model} | Distance: {self.distance} km | Fare: ₹{self.fare}"

d = Driver("Rajesh", "(Toyota Innova)", 4.5)
r = Rider("Amit", "Mumbai", "Pune")

r1 = Ride(d, r)
r1.calculate_fare(15)

r1.ride_info()
````
